```python
import requests # 웹 페이지의 HTML을 가져오는 모듈
from bs4 import BeautifulSoup # HTML 을 파싱하는 모듈

# HTML 파싱 : 텍스트 형태로 HTML 코드를 분석해 객체로 만든 뒤 검색하거나 편집할 
# 수 있도록  만드는 작업
# bs4 는 BeautifulSoup 라이브러리

# bs4는 Beau웹 페이지로 가져온 뒤 BeautifulSoup 객체로 만듦
response = requests.get('http://pythondojang.bitbucket.io/weather/observation/currentweather.html')
# requests.get 에 URL 을 넣으면 Response 응답 객체가 나옴
soup = BeautifulSoup(response.content, 'html.parser')
# BeautifulSoup 클래스에 응답객체의 content 속성과 'html.parser' 넣
# content 속성에는 텍스트 형태의 HTML이 들어있고
# 파이썬의 html.parser 모듈을 사용해서 파싱하도록 설정 
# BeautifulSoup 클래스로 만든 soup 객체로 태크를 찾음

table = soup.find('table', {'class':'table_develop3'}) # <table class="table_develop3"> 을 찾음
# HTML의 class 속성이 table_develop3 인 <table> 태그를 찾음
data = [] # 데이터를 저장할 리스트를 생성

for tr in table.find_all('tr'): # 모든 <tr> 태그를 찾아서 반복 (각 지점의 데이터를 가져옴)
    tds = list(tr.find_all('td')) # 모든 <td> 태그를 찾아서 리스트로 만듦
                                  # ( 각 날씨 값을 리스트로 만듦 )
    for td in tds: # <td> 태그 리스트 반복(각 날씨 값을 가져옴)
        if td.find('a'): # <td> 태그 안에 <a> 태그가 있으면(지점인지 확인)
            point = td.find('a').text  # <a> 태그 안에서 지점을 가져옴
            # td.find('a').text 와 같이 <a> 태그의 text 속성에서 지점을 가져옴 (<a>서울</a> 여기서 서울이 텍스트))
            temperature = tds[5].text # <td>태그 리스트의 여섯 번째(인덱스 5)에서 기온을 가져옴
            humidity = tds[9].text # <td> 태그 리스트의 열 번째(인덱스 9)에서 습도를 가져옴
            data.append([point, temperature, humidity]) # data 리스트에 지점, 기온, 습도 추가
                
data # data 표셔, 주피터 노트북에서는 print를 사용하지 않아도 변수의 값이 표시됨
```

```python
with open('weather.csv', 'w') as file: # weather.csv 파일을 쓰기 모드로 열기
    file.write('point,temperature,humidity\n') # 컬럼 이름 추가
    # 여기컬럼이름 추가할 때 문자열 콤마 사이사이에 띄어쓰기 하면 안됨
    # csv 파일 컬럼 이름 영어로 지정
    # csv 컬럼 접근할 때 df.temperature 처럼 속성 깔끔하게 사용할 수 있음
    for i in data: # data 를 반복하면서 
        file.write('{0}, {1}, {2}\n'.format(i[0], i[1], i[2])) # 지점, 온도, 습도를 줄 단위로 저장
        
```

```python
# weather.csv 파일을 읽어서 pandas 의 DdataFrame 객체로 만듬
# (pandas 는 데이터를 처리할 떄 사용하는 패키지)

# %matplotlib inline을 설정하면 matplotlib.pyplot 의 slow 함수를 호출하지 않아도 
# 주피터 노트북 안에서 그래프가 표시됨
# 위에꺼 안하고 그래프 표시하려면 plt.show()
%matplotlib inline 
import pandas as pd # 데이터를 저장하고 처리하는 패키지
import matplotlib as mpl # 그래프를 그리는 패키지
import matplotlib.pyplot as  plt  # 그래프를 그리는 패키지

# csv 파일을 읽어서  DataFrame 객체로 만듦. 인덱스 컬럼은 point로 설정
# html 가져온 페이지는  <meta http-equiv="Content-Type" content="text/html; charset=euc-kr" />
# 로 euc-kr 인코딩으로 만들어져 있기 때문에
# csv 파일도 euc-kr 로 인코딩  설정해서 파일을 읽어야 함
df = pd.read_csv('weather.csv', index_col = 'point', encoding = 'euc-kr')
# index_col 에 point를 지정해서 각 지점(point)이 맨 왼쪽에 오도록 설정
# index_col을 지정하지 않으면 맨 왼쪽 컬럼에는 0, 1, 2 처럼 숫자 인덱스가 들어감
df # DataFrame 으로 출력 

# 주피터 노트북은 DataFrame 을 표 형태로 보여줌
# DataFrame 은 엑셀 형태의  자료형인데 컬럼(열)과 로우(행)으로 구성됨
# csv 파일 내용을 별도의 설정없이 DataFrame으로 가져옴
# csv 파일 첫줄에는 컬럼이름이 있어야 함

```

```python
# 특별시 광역시만 모으기
# 특별시, 광역시만 모아서 따로 DataFrame 객체 만듦
city_df = df.loc[['서울', '인천','대전', '대구', '광주', '부산','울산']]
# DataFrame 의  loc 속성을 이용하면 특정 인덱스의 데이터만 가져올 수  있음
# DataFrame 객체.loc['인덱스 '] 인덱스 하나만 지정하면 해당 인덱스의 데이터만 가져오고
# DataFrame 객체.loc[['인덱스', '인덱스']] 인덱스 여러개를 리스트 형태로 지정하면 DataFrame 형태로 가져옴
# 이때 인덱스 문자열로 지정

city_df # city_df 출력
```

```python
# 특별시, 광역시만 모은  표로 그래프 그리기
# window 한글 폰트 설정

# matplotlib 는 기본적으로 그래프에서 한글 표시가 안되서 굴림폰트 정보를 가져온뒤
# get_name 으로 폰트 이름을 얻음
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name) # 폰트 설정

# 차트 종료, 제목, 차트 크기, 범례, 폰트 크기 설정
ax = city_df.plot(kind='bar', title='날씨', figsize =(12, 7), legend=True, fontsize=12 )
 # plot 메서드로 그래프 그리기
# plot 메서드를 사용하면 축 서브플롯(AxesSubplot) 객체가 나옴
ax.set_xlabel('도시', fontsize=12) # x 축 정보 표시
# set_xlabel, set_ylabel 메서드 사용
ax.set_ylabel('기온/습도', fontsize=12) # y 축 정보 표시
# legend 메서드에 범례를 리스트 형태로 넣음
ax.legend(['기온', '습도'], fontsize=12) # 범례 지정


```
