# 국가별로 데이터 정리하기

```python
team_list = df.set_index('Team').index # 전체 데이터 DataFrame 의
# 인덱스를 팀명으로 하고 그 팀명을  리스트 형식으로 저장
team_list # 팀명 리스트 출력
```

## 중복 국가 제거

```python
team = [] # 팀명 리스트 중 중복 제거된 팀명 저장할 빈 리스트 생성
for i in team_list: # 팀 리스트 요소 한개씩 변수에 대입
    if i not in team: # 빈 리스트에 변수값 없으면 
        team.append(i) # 중복 없이 새로운 팀명이니까 team리스트 요소로추가
        
team    # 중복 제거된 팀명 리스트 출력 
```

### data 접근 key를 나라이름으로 하려고 dictionary 자료형으로 만듦

```python
team_summary = {}
for i in range(len(team)): # 요소 개수 만큼 반복
    team_summary[team[i]] = df.set_index('Team').loc[team[i]].mean() #
    # 국가이름을 dictionary key로 하고  값을 국가별 데이터 평균 데이터 로 넣음
team_summary    
```

```python
team_summary['Korea Republic'] # dictionary key 값을 Korea Republic 로 해서  검색
```

### 파울을 가장 많이 한 나라 

```python
fouls = {}
goals = {}
possession = {}
distance = {}
yellow_and_red = {}
pass_accuracy = {}

for i in range(len(team)): # 팀 개수 만큼 반복
    fouls[team[i]] = team_summary[team[i]][13] # team_summary dictionary 키 값이
    # 나라이름이고 13번쨰 값 -> 파울 값  파울 dictionary 에 추가
    goals[team[i]] = team_summary[team[i]][0]
    possession[team[i]] = team_summary[team[i]][1]  
    pass_accuracy[team[i]] = team_summary[team[i]][9]  
    distance[team[i]] = team_summary[team[i]][11]
    yellow_and_red[team[i]] = team_summary[team[i]][15]
    
fouls
```

## 파울 많이 한 나라 그래프 그리기

```python
from pandas import Series, DataFrame
# dictionary 자료형 형태를 DataFrame 형태로 변환 
fouls_df = pd.DataFrame.from_dict(fouls, orient='index', columns=['fouls'])
fouls_df
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt



font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name) # 폰트 설정

ax = fouls_df.plot(kind='bar', title='국가별 파울 값', figsize=(12, 9), legend=True, fontsize=12)

ax.set_xlabel('국가', fontsize=12)
ax.set_ylabel('파울', fontsize=12)
ax.legend(['국가','파울'], fontsize=12)
```

```python
max(fouls, key=fouls.get) # dictionary 형태에서 max구한건데 DataFrame에서는 모르겟넴
```

```python
'가장 골 많이 넣은 나라 : {}'.format(max(goals, key=goals.get))
```

```python
'가장 점유율 높은 나라 : {}'.format(max(possession, key=possession.get))
```

```python
'가장 패스 정확도 높은 나라 : {}'.format(max(pass_accuracy, key=pass_accuracy.get))
```

```python
'가장 뛴거리가 많은 나라 : {}'.format(max(distance, key=distance.get))
```

```python
'가장 경고 & 퇴장 많이 받은 나라 : {}'.format(max(yellow_and_red, key=yellow_and_red.get))
```

## 가장 아름다운 축구를 한 나라
### 골은 +2, 점유율 +0.1, 패스정확도 +0.15, 뛴거리 +0.005, 경고 및 퇴장 -1, 가중치를 부여해서 가장 높은 나라 뽑기

```python
score = {} # 가중치 부여해서 계산한 결과값 저장할 빈 dictionary 생성
for i in range(len(team)):
        score[team[i]] = team_summary[team[i]][0]*2 + team_summary[team[i]][1]*0.1 \
                + team_summary[team[i]][9]*0.15 \
                        + team_summary[team[i]][11] *0.005 - team_summary[team[i]][15]
score
```

```python
'가장 아름다운 축구를 한 나라 : {}'.format(max(score, key=score.get))
```

```python
'가장 dirty 한 축구를 한 나라 : {}'.format(min(score, key=score.get))
```
